```python
import pandas as pd
df=pd.read_csv(r"C:\Users\Dell\Desktop\python\New folder\placement.csv")
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["cgpa"],df["package"])
plt.xlabel("CGPA")
plt.ylabel("package")
```

```python
x=df[['cgpa']] #independent feature / i/p
y=df[['package']] #independent feature / i/p
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x, y, test_size=0.20,random_state=39)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
print(dir(model))
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.predict([[6.89]]),model.predict([[5.66]])
```

```python
y_pred_test=model.predict(X_test)
y_pred_test,y_test
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred_test)
```
